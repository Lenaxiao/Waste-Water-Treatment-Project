```python
import pyowm
owm = pyowm.OWM('c0a3763a4ae7f31677879de49df9de49')
```

```python
import pandas as pd
data = pd.read_csv('T_UWWTPS.csv')
df_toclean = pd.DataFrame(data = {'Latitude': data['uwwLatitude'], 'Longitude': data['uwwLongitude']})
df_no_missing = df_toclean.dropna()
```

```python
df_no_missing
```

```python
df = df_no_missing.drop(df_no_missing.index[[3567]])
df
```

```python
lat1 = []
lon1 = []
lat1 = df_no_missing['Latitude'][0:2000].tolist()
lon1 = df_no_missing['Longitude'][0:2000].tolist()
location1 = pd.DataFrame({'Latitude': lat1, 'Longitude': lon1})
location1
```

```python
temp1 = []
for index, row in location1.iterrows():
    obs1 = owm.weather_at_coords(float(row['Latitude']),float(row['Longitude']))
    a1 = obs1.get_weather()
    a1.get_temperature('celsius')
    temp1.append(a1.get_temperature('celsius')['temp'])
```

```python
location1['temperature'] = temp1
location1
```

```python
location1.to_csv('location1.csv')
```
